```python
import cv2
import os
import random
```

```python
def showcase(dir_p):
    cases = os.listdir(dir_p)
    show_cases = random.choices(cases, k=10)
    fig, axes = plt.subplots(ncols=5, nrows=2, figsize=(25,10))
    axes = axes.flatten()
    for idx, case in enumerate(show_cases):
        img = cv2.imread(os.path.join(dir_p, case))
        axes[idx].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        axes[idx].axis('off')
    fig.tight_layout()
    plt.savefig('showheat.png', dpi=300)
```

```python
showcase('e:/New folder/')
```
